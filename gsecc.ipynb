```python
def plot_grupo_instrumentos(df_pz, df_rain, instrumentos, nombre_grupo, fecha_inicio, fecha_fin):
    """
    Genera una grÃ¡fica para un grupo de instrumentos piezomÃ©tricos
    
    Args:
        df_pz: DataFrame con datos piezomÃ©tricos (debe tener: id_instrumento, fecha_hora, elevacion_piezometrica)
        df_rain: DataFrame con datos de precipitaciÃ³n (debe tener: fecha_hora, rain_mm_tot)
        instrumentos: Lista de cÃ³digos de instrumentos a graficar
        nombre_grupo: Nombre descriptivo del grupo (para el tÃ­tulo)
        fecha_inicio: Fecha de inicio del rango
        fecha_fin: Fecha de fin del rango
    
    Returns:
        fig: Figura de matplotlib generada
    """
    
    fig, ax1 = plt.subplots(figsize=(16, 8))
    plt.style.use('bmh')
    
    # -----------------------------------
    # Filtrar datos del grupo
    # -----------------------------------
    df_grupo = df_pz[df_pz['id_instrumento'].isin(instrumentos)].copy()
    
    if df_grupo.empty:
        print(f"âš ï¸ No hay datos para el grupo '{nombre_grupo}'")
        plt.close(fig)
        return None
    
    # Filtrar por rango de fechas
    fecha_inicio_dt = pd.to_datetime(fecha_inicio)
    fecha_fin_dt = pd.to_datetime(fecha_fin)
    df_grupo = df_grupo[(df_grupo['fecha_hora'] >= fecha_inicio_dt) & 
                        (df_grupo['fecha_hora'] <= fecha_fin_dt)]
    
    if df_grupo.empty:
        print(f"âš ï¸ No hay datos en el rango de fechas para '{nombre_grupo}'")
        plt.close(fig)
        return None
    
    # -----------------------------------
    # Graficar cada instrumento
    # -----------------------------------
    colores = ['#00008B', '#DC143C', '#228B22', '#FF8C00', '#9370DB', 
               '#20B2AA', '#FF1493', '#8B4513', '#4682B4', '#DAA520']
    
    lines = []
    labels = []
    
    for i, instrumento in enumerate(instrumentos):
        df_inst = df_grupo[df_grupo['id_instrumento'] == instrumento]
        
        if df_inst.empty:
            print(f"  âš ï¸ Sin datos para {instrumento}")
            continue
        
        # Datos vÃ¡lidos
        datos_validos = df_inst[['fecha_hora', 'elevacion_piezometrica']].dropna()
        
        if datos_validos.empty:
            print(f"  âš ï¸ Sin datos vÃ¡lidos para {instrumento}")
            continue
        
        # Graficar serie
        color = colores[i % len(colores)]
        serie, = ax1.plot(
            datos_validos['fecha_hora'],
            datos_validos['elevacion_piezometrica'],
            color=color,
            label=instrumento,
            linewidth=2,
            marker='o' if len(datos_validos) == 1 else None,
            markersize=6,
            zorder=3,
            alpha=0.8
        )
        lines.append(serie)
        labels.append(instrumento)
        print(f"  âœ“ {instrumento}: {len(datos_validos)} puntos graficados")
    
    if not lines:
        print(f"âš ï¸ NingÃºn instrumento del grupo '{nombre_grupo}' tiene datos vÃ¡lidos")
        plt.close(fig)
        return None
    
    # -----------------------------------
    # Configurar eje Y
    # -----------------------------------
    ax1.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))
    ax1.set_ylabel('ElevaciÃ³n PiezomÃ©trica (msnm)', fontsize=14, fontweight='bold')
    ax1.tick_params(axis='y', labelsize=12, labelcolor='black')
    
    # Ajustar lÃ­mites Y con margen
    datos_validos_grupo = df_grupo[['fecha_hora', 'elevacion_piezometrica']].dropna()
    if not datos_validos_grupo.empty:
        y_min = datos_validos_grupo['elevacion_piezometrica'].min()
        y_max = datos_validos_grupo['elevacion_piezometrica'].max()
        if pd.notna(y_min) and pd.notna(y_max):
            rango = y_max - y_min
            margen = max(0.5, rango * 0.15)  # 15% del rango o mÃ­nimo 0.5
            ax1.set_ylim(y_min - margen, y_max + margen)
    
    # -----------------------------------
    # Eje secundario (PrecipitaciÃ³n)
    # -----------------------------------
    if df_rain is not None and not df_rain.empty:
        df_rain_filtrado = df_rain.copy()
        
        # Filtrar precipitaciÃ³n por fechas
        df_rain_filtrado = df_rain_filtrado[
            (df_rain_filtrado['fecha_hora'] >= fecha_inicio_dt) & 
            (df_rain_filtrado['fecha_hora'] <= fecha_fin_dt)
        ]
        
        if not df_rain_filtrado.empty and 'rain_mm_tot' in df_rain_filtrado.columns:
            # Calcular ancho de barras
            fecha_min = df_grupo['fecha_hora'].min()
            fecha_max = df_grupo['fecha_hora'].max()
            rango_dias = max(1, (fecha_max - fecha_min).days)
            
            if fecha_min == fecha_max:
                ancho_barra = 0.008
            elif rango_dias <= 10:
                ancho_barra = 0.015
            else:
                ancho_barra = max(0.035, rango_dias / 100 * 0.2)
            
            # Crear eje secundario
            ax2 = ax1.twinx()
            ax2.bar(
                df_rain_filtrado['fecha_hora'],
                df_rain_filtrado['rain_mm_tot'],
                label='PrecipitaciÃ³n',
                color='#009ACD',
                alpha=0.4,
                width=ancho_barra,
                zorder=1
            )
            ax2.set_ylabel('PrecipitaciÃ³n (mm/dÃ­a)', color='black', fontsize=14, fontweight='bold')
            ax2.tick_params(axis='y', labelcolor='black', labelsize=12)
            ax2.grid(False)
            
            # Ajustar lÃ­mite Y de precipitaciÃ³n
            max_precip = df_rain_filtrado['rain_mm_tot'].max()
            ax2.set_ylim(0, max_precip * 1.2)  # 20% de margen
        
            
            # AÃ±adir a la leyenda
            lines.append(Line2D([], [], color='#009ACD', linewidth=4, alpha=0.4))
            labels.append('PrecipitaciÃ³n')
    
    # -----------------------------------
    # ConfiguraciÃ³n del eje X
    # -----------------------------------
    fecha_min = df_grupo['fecha_hora'].min()
    fecha_max = df_grupo['fecha_hora'].max()
    
    if fecha_min == fecha_max:
        fecha_min -= pd.Timedelta(days=1)
        fecha_max += pd.Timedelta(days=1)
    
    ax1.set_xlim([fecha_min, fecha_max])
    
    rango_dias = max(1, (fecha_max - fecha_min).days)
    
    if rango_dias > 365: 
        intervalo = 60
    elif rango_dias > 180: 
        intervalo = 30
    elif rango_dias > 90: 
        intervalo = 15
    elif rango_dias > 31: 
        intervalo = 7
    else: 
        intervalo = 1
    
    ax1.xaxis.set_major_locator(mdates.DayLocator(interval=intervalo))
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
    ax1.tick_params(axis='x', rotation=45, labelsize=11, labelcolor='black')
    
    # -----------------------------------
    # TÃ­tulo y formato
    # -----------------------------------
    ax1.set_title(f'SECCIÃ“N GEOTÃ‰CNICA {nombre_grupo} - DRT NIVELES PIEZOMÃ‰TRICOS',
                  fontsize=18,
                  fontweight='bold',
                  pad=20)
    
    ax1.grid(True, linestyle='--', color='gray', alpha=0.5, zorder=0)
    ax1.set_facecolor('white')
    
    for spine in ax1.spines.values():
        spine.set_edgecolor('silver')
        spine.set_linewidth(1.5)
    
    # -----------------------------------
    # Leyenda
    # -----------------------------------
    ax1.legend(lines, labels, 
               loc='lower center', 
               bbox_to_anchor=(0.5, -0.25), 
               ncol=min(6, len(lines)), 
               fontsize=11, 
               frameon=True, 
               facecolor='white', 
               edgecolor='silver',
               shadow=True)
    
    fig.tight_layout()
    print(f"âœ“ GrÃ¡fico generado para grupo '{nombre_grupo}'")
    return fig


# ========================================================================
# FUNCIÃ“N PRINCIPAL: Generar grÃ¡ficas para todos los grupos
# ========================================================================

def generar_graficas_por_grupos(df_pz, df_rain, diccionario, fecha_inicio, fecha_fin, 
                                 guardar=True, mostrar=True):
    """
    Genera una grÃ¡fica por cada grupo del diccionario
    
    Args:
        df_pz: DataFrame con datos piezomÃ©tricos
        df_rain: DataFrame con datos de precipitaciÃ³n
        diccionario: Dict con grupos de instrumentos {nombre: [lista_instrumentos]}
        fecha_inicio: Fecha inicial del rango
        fecha_fin: Fecha final del rango
        guardar: Si True, guarda las grÃ¡ficas como PNG
        mostrar: Si True, muestra las grÃ¡ficas en pantalla
    
    Returns:
        List de figuras generadas
    """
    
    figuras = []
    
    print("="*70)
    print("GENERANDO GRÃFICAS POR GRUPOS")
    print("="*70)
    
    for nombre_grupo, instrumentos in diccionario.items():
        print(f"\nðŸ“Š Procesando Grupo '{nombre_grupo}':")
        print(f"   Instrumentos: {', '.join(instrumentos)}")
        
        fig = plot_grupo_instrumentos(
            df_pz=df_pz,
            df_rain=df_rain,
            instrumentos=instrumentos,
            nombre_grupo=nombre_grupo,
            fecha_inicio=fecha_inicio,
            fecha_fin=fecha_fin
        )
        
        if fig is not None:
            figuras.append((nombre_grupo, fig))
            
            if guardar:
                filename = f'grafico_grupo_{nombre_grupo}.png'
                fig.savefig(filename, dpi=300, bbox_inches='tight')
                print(f"   ðŸ’¾ Guardado: {filename}")
            
            if mostrar:
                plt.show()
            else:
                plt.close(fig)
        else:
            print(f"   âŒ No se pudo generar grÃ¡fico para grupo '{nombre_grupo}'")
    
    print("\n" + "="*70)
    print(f"âœ“ PROCESO COMPLETADO: {len(figuras)} grÃ¡ficas generadas")
    print("="*70)
    
    return figuras


if __name__ == "__main__":
    

    
    # 6. Definir rango de fechas
    fecha_inicio = '2024-06-01'
    fecha_fin = '2025-12-21'
    
    # 7. Generar grÃ¡ficas
    figuras = generar_graficas_por_grupos(
        df_pz=df_pz,
        df_rain=df_rain,
        diccionario=diccionario,
        fecha_inicio=fecha_inicio,
        fecha_fin=fecha_fin,
        guardar=True,      # Guardar como PNG
        mostrar=True       # Mostrar en pantalla
    )
    
    print(f"\nâœ… Total de grÃ¡ficas generadas: {len(figuras)}")
```
